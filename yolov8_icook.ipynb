## 切出bounding box


```python
import cv2
import numpy as np
from tensorflow.keras.applications.resnet50 import preprocess_input
```

```python
def crop_images_with_bboxes(images, bboxes):
    cropped_images = []
    for image, bbox in zip(images, bboxes):
        image_np = image.cpu().numpy().transpose(1, 2, 0)  # Convert PyTorch tensor to numpy array
        x1, y1, x2, y2 = map(int, bbox[:4])  # Extract bounding box coordinates
        cropped_image = image_np[y1:y2, x1:x2]
        cropped_images.append(cropped_image)
    return cropped_images

def preprocess_for_resnet(cropped_images):
    processed_images = []
    for img in cropped_images:
        img = cv2.resize(img, (224, 224))  # Resize to 224x224
        img = preprocess_input(img)       # Normalize the image
        processed_images.append(img)
    return np.array(processed_images)
```
